Use the given file named cricket.csv and perform the following operations:
1. Read the file in DataFrame
2. List the name of cricketer and their respective runs
3. Find total wickets taken by them
4. Find average of catches taken
5. Find the name of wicketkeeper
6. print the name of bowler who played highest number of matches
7. Find average of all the bowlers
8. Find name of the bowler with least bowling average (wickets per match).
9. Print information about all players whose catches per match ratio is greater than 0.5
store it in a csv file.
10. Draw the bar chart of playr name against their runs.
11. Sort and print information about players by ascending order of runs
12. print the names of players whose wickets are greater than matches

```python
#Read the file in DataFrame
import pandas as pd
data1 = pd.read_csv('cricket.csv')
```

```python
data1.head()
```

```python
#List the name of cricketer and their respective runs
data1[["name","runs"]]
```

```python
#Find total wickets taken by them
data1[["name","wickets"]]
```

```python
#Find average of catches taken
data1.loc[:,'catches'].mean()
```

```python
#Find the name of wicketkeeper
data1[data1['stumpings']!=0]['name']
```

```python
#print the name of bowler who played highest number of matches
data1_bowlers=data1[data1['wickets']!=0]
data1_bowlers[data1_bowlers['matches']==data1_bowlers.loc[:,'matches'].max()]['name']
```

```python
# Find average of all the bowlers
# As the number of runs conceded by the bowler have not been given,
# the average has been calculated as wickets per match

data1_bowlers.loc[:,'bowling_average']=data1_bowlers.loc[:,'wickets'] / data1_bowlers.loc[:,'matches']
data1_bowlers
```

```python
# Find name of the bowler with least bowling average (wickets per match).
data1_bowlers[data1_bowlers['bowling_average'] == data1_bowlers['bowling_average'].min()]['name']
```

```python
# Print information about all players whose catches per match ratio is greater than 0.5
# store it in a csv file.
data1_bowlers[data1_bowlers['bowling_average'] > 0.5].to_csv("Bowling_average_lessthan_0pt5.csv")
```

```python
#Draw the bar chart of playr name against their runs.
data1[['name','runs']]
```

```python
#Sort and print information about players by ascending order of runs
data1.sort_values(["runs"],ignore_index=True)
```

```python
#print the names of players whose wickets are greater than matches
data1[(data1['wickets'])>(data1['matches'])]['name']
```

posted on Facebook page.

1. Count how many number of posts are paid and unpaid?
 2. Display only Page type, Month, likes, comments and shares.
 3. Find average number of likes to the post in September month.
 4. Print the information about the posts which are links.
 5. How many total shares are present for the video posts?
 6. In which hour of the day maximum posts are posted?
 7. In which day of October, maximum number of interactions are found?

```python
# Count how many number of posts are paid and unpaid?
import pandas as pd
data2 = pd.read_csv("./fb.csv")
data2
```

```python
print("Number of paid posts")
data2['Paid']
```

```python
#Display only Page type, Month, likes, comments and shares.
data2[['Type','Post Month','like','comment','share']]
```

```python
# Find average number of likes to the post in September month.
data2_Sept = data2[data2['Post Month'] == 9]
print("Average number of likes in september are", data2_Sept['like'].mean())
```

```python
# Print the information about the posts which are links.
data2[data2['Type'] == 'Link']
```

```python
# How many total shares are present for the video posts?
data2_video = data2[data2['Type'] == 'Video']
print("Total shares for video posts", data2_video['share'].sum())
```

```python
# In which hour of the day maximum posts are posted?
data2['Post Hour'].value_counts().head(1)
```

```python
#In which day of October, maximum number of interactions are found?
data2_October = data2[data2['Post Month'] == 10]
max_interactions = data2_October['Total Interactions'].max()
data2[data2['Total Interactions'] == max_interactions]['Post Weekday']
```

```python
# Print the information of the posts where likes are above 200 and shares are above 30.
data2[(data2['like'] > 200) & (data2['share'] > 30)]
```

```python
#Which kind of posts are posted maximum number of times in November month?
data2_November = data2[data2['Post Month'] == 11]
data2_November['Type'].value_counts().head(1)
```

```python
#Find average and median of interactions for the paid (1) posts
data2_paid = data2[data2['Paid'] == 1.0]
print("Average: ",data2_paid['Total Interactions'].mean())
print("Median: ",data2_paid['Total Interactions'].median())
```
